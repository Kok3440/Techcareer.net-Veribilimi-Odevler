```python
# 1. 1. Numpy ile Matris İşlemleri
# - 5x5 boyutunda rastgele (0-100 arasında) tam sayılardan oluşan bir matris oluşturun.
# - Bu matrisin:
 # • Ortalama, standart sapma ve varyansını hesaplayın.
 # • En büyük ve en küçük değerlerini bulun.
 # • Köşegenindeki elemanların toplamını bulun.

import numpy as np
import random as rn

matris = np.random.randint(0, 100, (5, 5))
print("Matris:")
print(matris)

# Calculate statistics
mean = np.mean(matris)
std_dev = np.std(matris)
variance = np.var(matris)
max_val = np.max(matris)
min_val = np.min(matris)
diagonal_sum = np.trace(matris)

print("\nİstatistikler:")
print(f"Ortalama: {mean}")
print(f"Standart Sapma: {std_dev}")
print(f"Varyans: {variance}")
print(f"En büyük değer: {max_val}")
print(f"En küçük değer: {min_val}")
print(f"Köşegen elemanları toplamı: {diagonal_sum}")
```

```python
#2. 2. Numpy ile Veri Simülasyonu
#- 1000 adet öğrencinin sınav puanlarını (0-100 arası, normal dağılımdan gelen) simüle edin.
#- Ortalama, medyan ve standart sapmayı hesaplayın.
#- 50’den düşük alan kaç öğrenci olduğunu bulun.

ogrenci_puanlari = np.random.normal(50, 10, 1000)
ortalama_puan = np.mean(ogrenci_puanlari)
medyan_puan = np.median(ogrenci_puanlari)
standart_sapma = np.std(ogrenci_puanlari)
kalan_ogrenci=ogrenci_puanlari[ogrenci_puanlari<50]
print("Öğrenci puanları:")
print(ogrenci_puanlari)
print("\nİstatistikler:")
print(f"Ortalama: {ortalama_puan}")
print(f"Medyan: {medyan_puan}")
print(f"Standart Sapma: {standart_sapma}")
print(f"50'den düşük alan öğrenci sayısı: {len(kalan_ogrenci)}")
```

```python
#• Görevler:
#3. 1. Veri setini Pandas DataFrame olarak oluşturun.
#4. 2. Her ders için ortalama puanı bulun.
#5. 3. En yüksek matematik notunu alan öğrenciyi bulun.
#6. 4. Her öğrencinin not ortalamasını hesaplayan yeni bir sütun ekleyin.
#7. 5. Bölümlere göre gruplayarak her bölümün ortalama başarılarını hesaplayın.
#8. 6. Ortalaması 70’in üzerinde olan öğrencileri filtreleyin.
from google.colab import files
import pandas as pd

uploaded = files.upload()

# Assuming the uploaded file is a CSV, replace "your_dataset.csv" with the actual filename
# If it's a different format, use the appropriate pandas read function (e.g., pd.read_excel)
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

# Display the first few rows of the DataFrame to verify
print("DataFrame created from the dataset:")
display(df.head())
```

```python
# 2. Her ders için ortalama puanı bulun.
average_scores = df[['Matematik', 'Fizik', 'Kimya']].mean()
print("\nHer ders için ortalama puanlar:")
print(average_scores)
```

```python
# 3. En yüksek matematik notunu alan öğrenciyi bulun.
highest_math_student = df.loc[df['Matematik'].idxmax()]
print("\nEn yüksek matematik notunu alan öğrenci:")
print(highest_math_student)
```

```python
#  4. Her öğrencinin not ortalamasını hesaplayan yeni bir sütun ekleyin.
df['Ortalama Not'] = df[['Matematik', 'Fizik', 'Kimya']].mean(axis=1)
print("\nDataFrame with student averages:")
display(df.head())
```

```python
# 5. Bölümlere göre gruplayarak her bölümün ortalama başarılarını hesaplayın.
department_averages = df.groupby('Bölüm')['Ortalama Not'].mean()
print("\nBölümlere göre ortalama başarılar:")
print(department_averages)
```

```python
# 6. Ortalaması 70’in üzerinde olan öğrencileri filtreleyin.
students_above_70 = df[df['Ortalama Not'] > 70]
print("\nOrtalaması 70'in üzerinde olan öğrenciler:")
display(students_above_70)
```
